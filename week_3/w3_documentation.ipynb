Inputs

Quiz 1 mark
Quiz 2 mark
Quiz 3 mark
Continue choice (Y/N)

Process

Read the three quiz marks.
Calculate the average.
Compare the average with 50.
Repeat if user selects Y.

Outputs

Average mark
PASS or FAIL message
Program Ended message

Pseudocode
START

Set choice = "y"

WHILE choice = "y"

    Input quiz_1
    Input quiz_2
    Input quiz_3

    average = (quiz_1 + quiz_2 + quiz_3) / 3

    Display average

    IF average >= 50 THEN
        Display "PASS"
    ELSE
        Display "FAIL"
    ENDIF

    Input choice

ENDWHILE

Display "Program Ended"

END

Algorithm Structure
Sequence: Input marks → Calculate average → Display result
Selection: if average >= 50
Iteration: while choice == "y"
